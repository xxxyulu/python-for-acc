```python
data=pd.read_csv('链家房源区位数据.csv')
data
```

```python
data['地理位置']=data['位置1']+data['位置2']+data['小区']
data
```

```python
data1=data.sort_values(by=['价格'],ascending=[False]).head(100)
data1
```

```python
#获取坐标函数
def getlnglat(address):
    url = 'http://api.map.baidu.com/geocoding/v3/'#网上最新版本V3
    output = 'json'
    ak = '5qbQQfFhNVyCiw0mOTEvlHxiolO3oKAV' # 百度地图ak，具体申请自行百度，提醒需要在“控制台”-“设置”-“启动服务”-“正逆地理编码”，启动
    address = quote(address) # 由于本文地址变量为中文，为防止乱码，先用quote进行编码 key=urllib.request.quote(keyworld)
    uri = url + '?' + 'address=' + address + '&output=' + output + '&ak=' + ak
    req = urlopen(uri)
    res = req.read().decode()
    temp = json.loads(res)
    lat = temp['result']['location']['lat']
    lng = temp['result']['location']['lng']
    return lat,lng # 纬度 latitude ， 经度 longitude ，
```

```python
#在原有excel表的基础上增加经纬度信息
for indexs in data1.index:
    get_location = getlnglat(data1.loc[indexs,'地理位置'])
    get_lat = get_location[0]
    get_lng = get_location[1]
    data1.loc[indexs,'纬度'] = get_lat
    data1.loc[indexs,'经度'] = get_lng
    
data1
```

```python
import webbrowser as wb
import pandas as pd
import folium
```

```python
#百度坐标转换为高德坐标
import math
def bdToGaoDe(lon,lat):
    PI = 3.14159265358979324 * 3000.0 / 180.0
    x = lon - 0.0065
    y = lat - 0.006
    z = math.sqrt(x * x + y * y) - 0.00002 * math.sin(y * PI)
    theta = math.atan2(y, x) - 0.000003 * math.cos(x * PI)
    lon = z * math.cos(theta)
    lat = z * math.sin(theta)
    return lon,lat
```

```python
for indexs in data1.index:
    get_jw = bdToGaoDe(data1.loc[indexs,'经度'],data1.loc[indexs,'纬度'])
    get_lat = get_jw[1]
    get_lng = get_jw[0]
    data1.loc[indexs,'纬度'] = get_lat
    data1.loc[indexs,'经度'] = get_lng
    
data1
```
